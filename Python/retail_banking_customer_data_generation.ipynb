```python
import pandas as pd
import numpy as np

# -------------------------------
# Enterprise Retail Banking Dataset Generator
# -------------------------------

np.random.seed(42)

records = 100000

customer_ids = [f"CUST{100000+i}" for i in range(records)]

ages = np.random.randint(18, 75, records)

genders = np.random.choice(
    ["Male", "Female"],
    records,
    p=[0.52,0.48]
)

regions = np.random.choice(
    ["North","South","East","West"],
    records
)

city_map = {
    "North":["Delhi","Chandigarh","Lucknow"],
    "South":["Bengaluru","Hyderabad","Chennai"],
    "East":["Kolkata","Bhubaneswar","Patna"],
    "West":["Mumbai","Pune","Ahmedabad"]
}

cities = [
    np.random.choice(city_map[r])
    for r in regions
]

income = np.random.randint(200000,3000000,records)

income_category = pd.cut(
    income,
    bins=[0,500000,1000000,2000000,10000000],
    labels=["Low","Medium","High","Premium"]
)

occupations = np.random.choice(
    [
        "Salaried",
        "Business",
        "Professional",
        "Government",
        "Self Employed"
    ],
    records
)

customer_segment = np.random.choice(
    [
        "Mass",
        "Silver",
        "Gold",
        "Platinum"
    ],
    records,
    p=[0.45,0.30,0.18,0.07]
)

deposit_balance = np.random.randint(
    5000,
    15000,
    records
)

loan_amount = np.random.randint(
    1000,
    5000,
    records
)

products = np.random.randint(
    1,
    7,
    records
)

primary_product = np.random.choice(
    [
        "Savings Account",
        "Current Account",
        "Home Loan",
        "Personal Loan",
        "Credit Card",
        "Fixed Deposit"
    ],
    records
)

mobile_banking = np.random.choice(
    ["Yes","No"],
    records,
    p=[0.82,0.18]
)

internet_banking = np.random.choice(
    ["Yes","No"],
    records,
    p=[0.75,0.25]
)

credit_card = np.random.choice(
    ["Yes","No"],
    records,
    p=[0.55,0.45]
)

active_customer = np.random.choice(
    ["Yes","No"],
    records,
    p=[0.85,0.15]
)

customer_tenure = np.random.randint(
    1,
    21,
    records
)

monthly_transactions = np.random.randint(
    5,
    120,
    records
)

customer_value = np.where(
    deposit_balance>12000,
    "High",
    np.where(
        deposit_balance>8000,
        "Medium",
        "Low"
    )
)

revenue = (
    deposit_balance*0.02 +
    loan_amount*0.05 +
    products*500
).round(2)

cross_sell = np.where(
    products<=2,
    "High",
    np.where(
        products<=4,
        "Medium",
        "Low"
    )
)

df = pd.DataFrame({

    "Customer_ID":customer_ids,
    "Age":ages,
    "Gender":genders,
    "Region":regions,
    "City":cities,
    "Income":income,
    "Income_Category":income_category,
    "Occupation":occupations,
    "Customer_Segment":customer_segment,
    "Deposit_Balance":deposit_balance,
    "Loan_Amount":loan_amount,
    "Number_of_Products":products,
    "Primary_Product":primary_product,
    "Mobile_Banking":mobile_banking,
    "Internet_Banking":internet_banking,
    "Credit_Card":credit_card,
    "Active_Customer":active_customer,
    "Customer_Tenure":customer_tenure,
    "Monthly_Transactions":monthly_transactions,
    "Customer_Value":customer_value,
    "Revenue":revenue,
    "Cross_Sell_Opportunity":cross_sell

})

df.to_csv(
    "retail_banking_customer_dataset.csv",
    index=False
)

print("="*60)
print("Retail Banking Customer Dataset Generated Successfully")
print("="*60)

print(f"Customers              : {len(df):,}")
print(f"Deposit Portfolio      : ₹{df['Deposit_Balance'].sum():,.0f}")
print(f"Loan Portfolio         : ₹{df['Loan_Amount'].sum():,.0f}")
print(f"Average Deposit        : ₹{df['Deposit_Balance'].mean():,.2f}")
print(f"Average Loan           : ₹{df['Loan_Amount'].mean():,.2f}")
print(f"Active Customers       : {(df['Active_Customer']=='Yes').sum():,}")
print(f"Mobile Banking Users   : {(df['Mobile_Banking']=='Yes').sum():,}")
print(f"Internet Banking Users : {(df['Internet_Banking']=='Yes').sum():,}")
print("="*60)
```
